# Visualisation

Este cÃ³digo genera una grÃ¡fica de pares para los datos del conjunto de la base de datos Iris. Esta grÃ¡fica muestra la relaciÃ³n entre todas las caracterÃ­sticas del conjunto de datos en un solo lugar. La grÃ¡fica estÃ¡ coloreada por la especie, lo que permite ver cÃ³mo cada especie se comporta con respecto a las otras. El parÃ¡metro 'altura' controla el tamaÃ±o de la figura y el parÃ¡metro 'aspecto' controla el tamaÃ±o relativo de los ejes x e y.

https://seaborn.pydata.org/generated/seaborn.pairplot.html

```python
sns.pairplot(iris, hue='species', height=5, aspect=1);
```

```python
iris.hist(edgecolor='red', linewidth=3, figsize=(20,8));
plt.show();
```

```python
plt.figure(figsize=(20,8));
plt.subplot(2,2,1)
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='species', y='sepal_width', data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='species', y='petal_length', data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='species', y='petal_width', data=iris);
```

```python
iris.boxplot(by='species', figsize=(15,8));
```

```python
pd.plotting.scatter_matrix(iris, figsize=(15,10))
plt.show()
```
