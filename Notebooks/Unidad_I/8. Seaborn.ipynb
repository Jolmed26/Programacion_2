# Visualisation

Este código genera una gráfica de pares para los datos del conjunto de la base de datos Iris. Esta gráfica muestra la relación entre todas las características del conjunto de datos en un solo lugar. La gráfica está coloreada por la especie, lo que permite ver cómo cada especie se comporta con respecto a las otras. El parámetro 'altura' controla el tamaño de la figura y el parámetro 'aspecto' controla el tamaño relativo de los ejes x e y.

https://seaborn.pydata.org/generated/seaborn.pairplot.html

```python
sns.pairplot(iris, hue='species', height=5, aspect=1);
```

```python
iris.hist(edgecolor='red', linewidth=3, figsize=(20,8));
plt.show();
```

```python
plt.figure(figsize=(20,8));
plt.subplot(2,2,1)
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='species', y='sepal_width', data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='species', y='petal_length', data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='species', y='petal_width', data=iris);
```

```python
iris.boxplot(by='species', figsize=(15,8));
```
